```python
import pandas as pd

df = pd.read_csv(
    r"C:\Users\ankit\OneDrive\Documents\Desktop\Sales_Data_Analysis\data\sales_data.csv"
)

df.head()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.dropna(inplace=True)
```

```python
df['Total Sales'] = df['Quantity'] * df['Price']
df['Month'] = df['Order Date'].dt.month
```

```python
df['Total Sales'].sum()                      # total sales
df.groupby('Product')['Total Sales'].sum()   # product sales
df.groupby('Month')['Total Sales'].sum()     # monthly sales
df.groupby('Region')['Total Sales'].sum()    # region sales
```

```python
import matplotlib.pyplot as plt

df.groupby('Month')['Total Sales'].sum().plot(kind='bar')
plt.show()
```

```python
import matplotlib.pyplot as plt

monthly_sales = df.groupby('Month')['Total Sales'].sum()
monthly_sales.plot(kind='bar', color='skyblue')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

```python
product_sales = df.groupby('Product')['Total Sales'].sum()
product_sales.plot(kind='bar', color='orange')
plt.title("Sales by Product")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show()
```

```python
region_sales = df.groupby('Region')['Total Sales'].sum()
region_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title("Sales by Region")
plt.show()
```
